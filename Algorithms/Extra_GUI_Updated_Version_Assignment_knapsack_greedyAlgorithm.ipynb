```python
class Knapsack:
    """
    A class to represent a program of knapsack.

    Attributes
    ----------
    weight : list of integers
        a list of weights of items
    value : list of integers
        a list of values of items
    Mass : float
        capacity of knapsack
    types : string
        type of knapsak : 'zero one' or 'fractional'
    method : string
        method to fill knapsack 'max value' or 'least weight' or 'value per weight' or 'all methods'
    answer : float
        the variable used to calculate the result
    -------
    """
    #constructor
    def __init__(self,weight,value,Mass,types,method,answer=0):
        self.w=weight
        self.v=value
        self.M=Mass
        self.types=types
        self.method=method
        self.answer=answer
    #a function to calculate knapsack result using max value
    def max_value(self,w,v,M,t,method,answer):
        if t=='zero one': #case zero one knapsack
            while self.M>0 and len(self.w)>0: #check that mass of knapsack isn't equal 0 and there are still items to calculate
                maxobj=max(self.v) #get the item of maximum value
                maxindex=self.v.index(maxobj) #get the ndex of the item with maximum value
                if self.w[maxindex]<=self.M: # check that the weight of the item with maximum value is not greater than capacity
                    self.answer+=maxobj #add the value of the item
                    self.M-=self.w[maxindex] #subtract the weight of the item from the total capacity of knapsack
                #delete the value and the weight of the item from the lists after being calculated
                del self.v[maxindex]
                del self.w[maxindex]
            return 'answer of max value of zero one knapsack ={} '.format(self.answer) #return final result
        elif t=='fractional': #case fractional knapsack
            while self.M>0 and len(self.w)>0:
                maxobj=max(self.v)
                maxindex=self.v.index(maxobj)
                if self.w[maxindex]<=self.M:
                    self.answer+=maxobj
                    self.M-=self.w[maxindex]
    
                else:
                    #creating a list for value per weight of item
                    zipping=zip(self.v,self.w)
                    vperw=[i[0]/i[1] for i in zipping]
                    self.answer+=(self.M*vperw[maxindex]) #calculating the value of the chunck needed to fill the knapsack
                    break
                del self.v[maxindex]
                del self.w[maxindex]
            return 'answer of max value of fractional knapsack ={} '.format(self.answer)
        else:
            #incase invalid input
            return 'invalid input for type of knapsack'

    #a function to calculate knapsack result using least weight
    def least_weight(self,w,v,M,t,method,answer):
        if t=='zero one':
            while self.M>0 and len(self.w)>0:
                leastweight=min(self.w) #get the item of minimum weight
                minindex=self.w.index(leastweight) #index of item of minimum weight
                if self.w[minindex]<=self.M:
                    self.answer+=self.v[minindex]
                    self.M-=self.w[minindex]
                del self.v[minindex]
                del self.w[minindex]
            return 'answer of least weight of zero one knapsack ={} '.format(self.answer)
        elif t=='fractional':
            while self.M>0 and len(self.w)>0:
                leastweight=min(self.w)
                minindex=self.w.index(leastweight)
                if self.w[minindex]<=self.M:
                    self.answer+=v[minindex]
                    self.M-=self.w[minindex]

                else:
                    zipping=zip(self.v,self.w)
                    vperw=[i[0]/i[1] for i in zipping]
                    self.answer+=(self.M*vperw[minindex])
                    break
                del self.v[minindex]
                del self.w[minindex]
            return 'answer of least weight value of fractional knapsack ={} '.format(self.answer)
        else:
            return 'invalid input for type of knapsack'
        
    #a function to calculate knapsack result using maximum value per weight
    def value_per_weight(self,w,v,M,t,method,answer):
        if t=='zero one':
            zipping=zip(self.v,self.w)
            vperw=[i[0]/i[1] for i in zipping]
            while self.M>0 and len(self.w)>0:
                maxobj=max(vperw)
                maxindex=vperw.index(maxobj)
                if self.w[maxindex]<=self.M:
                    self.answer+=self.v[maxindex]
                    self.M-=self.w[maxindex]
                #delete the value and the weight and the value per weight of the item from the lists after being calculated
                del self.v[maxindex]
                del self.w[maxindex]
                del vperw[maxindex]
            return 'answer of value per weight of zero one knapsack ={} '.format(self.answer)
        elif t=='fractional':
            zipping=zip(self.v,self.w)
            vperw=[i[0]/i[1] for i in zipping]
            while self.M>0 and len(self.w)>0:
                maxobj=max(vperw)
                maxindex=vperw.index(maxobj)
                if self.w[maxindex]<=self.M:
                    self.answer+=self.v[maxindex]
                    self.M-=self.w[maxindex]
                else:
                    self.answer+=(self.M*(vperw[maxindex]))
                    break
                del self.v[maxindex]
                del self.w[maxindex]
                del vperw[maxindex]
            return 'answer of value per weight of fractional knapsack ={} '.format(self.answer)
        else:
            return 'invalid input for type of knapsack'
    
    #a function to call the suitable function from the previouse functions according to the input
    def knapsack_answer(self):
        answer=0
        if (len(self.w)!=len(self.v)): #check that every item has its own weight and value
            return 'length of weights list and values list are not equal'
        elif self.M<0: #check that the knapsack's capacity is valid
            return 'invalid input for capacity of knapsack'
       
        elif self.method=='max value': #incase user wants to calculate knapsack with max value method
            return self.max_value(self.w,self.v,self.M,self.types,self.method,self.answer)

        elif self.method=='least weight': #incase user wants to calculate knapsack with least weight method
            return self.least_weight(self.w,self.v,self.M,self.types,self.method,self.answer)
                
        elif self.method=='value per weight': #incase user wants to calculate knapsack with value per weight method
            return  self.value_per_weight(self.w,self.v,self.M,self.types,self.method,self.answer)
        
        elif self.method=='all methods': #incase user wants to calculate all methods for a specific type of knapsack
            tempv=self.v.copy()
            tempw=self.w.copy()
            temp2v=self.v.copy()
            temp2w=self.w.copy()
            temp1v=self.v.copy()
            temp1w=self.w.copy()
            if self.types=='zero one': #if user wanted to calculate all methods for zero one knapsack
                k1,k2,k3=Knapsack(tempw,tempv,self.M,self.types,'max value'),Knapsack(temp2w,temp2v,self.M,self.types,'least weight'),Knapsack(temp1w,temp1v,self.M,self.types,'value per weight')
                return k1.knapsack_answer(),k2.knapsack_answer(),k3.knapsack_answer()
            elif self.types=='fractional': #if user wanted to calculate all methods for fractional knacpsack
                k1,k2,k3=Knapsack(tempw,tempv,self.M,self.types,'max value'),Knapsack(temp2w,temp2v,self.M,self.types,'least weight'),Knapsack(temp1w,temp1v,self.M,self.types,'value per weight')
                return k1.knapsack_answer(),k2.knapsack_answer(),k3.knapsack_answer()
            else:
                return 'invalid input for type of knapsack'
        else: #incase user entered invalid input for the method of knapsack
            return 'invalid input for method of knapsack'
'''
#----------------------------------------------------------------------------------------------------------
#the test cases:

#input:
#w=[1,3,5,4,1,3,2]
#v=[5,10,15,7,8,9,4]
#Mass=15

#the right output for type=zero one , method=all methods: 'max value','least weight','value per weight':
#                                                             51            43               51
#the right output for type=fractional , method=all methods: 'max value','least weight','value per weight':
#                                                              47.25         46               51

z=1
while z:
    try:
        w=[float(i) for i in input('please enter weights list ').split()]
        v=[float(i) for i in input('please enter values list ').split()]
        M=float(input('please enter capacity of knapsack '))
        t=input('please enter type of knapsack : "zero one" or "fractional" ')
        method=input('please enter method : "max value" or "least weight" or "value per weight" or "all methods" ')
        knap=Knapsack(w,v,M,t,method)
        print(knap.knapsack_answer())
        z=int(input('enter 1 for another inputs and 0 to exit '))
    except:
        print('invalid input, please try again')
'''
help(Knapsack)
```
